3. Handle Missing Values

```python
 #Number of NAN values on each Column
print("Column 'Synonyms'    has     : (",data['Synonyms'].isnull().sum() ,") Null value")
print("Column 'Japanese'    has     : ( "  ,data['Japanese'].isnull().sum() ," ) Null value")
print("Column 'English'     has     : (",data['English'].isnull().sum() ,") Null value")
print("Column 'Premiered'   has     : (",data['Premiered'].isnull().sum() ,") Null value")
print("Column 'Broadcast'   has     : (",data['Broadcast'].isnull().sum() ,") Null value")
print("Column 'Genres'      has     : (",data['Genres'].isnull().sum() ,") Null value")
print("Column 'Demographic' has     : (",data['Demographic'].isnull().sum() ,") Null value")
```

```python
data.shape
```

```python
from sklearn.impute import SimpleImputer

# Separate numeric and non-numeric columns
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
non_numeric_columns = data.select_dtypes(exclude=['float64', 'int64']).columns

# Impute missing values for numeric columns with mean strategy
numeric_imputer = SimpleImputer(strategy='mean')
data[numeric_columns] = numeric_imputer.fit_transform(data[numeric_columns])

# Impute missing values for non-numeric columns with most frequent strategy
non_numeric_imputer = SimpleImputer(strategy='most_frequent')
data[non_numeric_columns] = non_numeric_imputer.fit_transform(data[non_numeric_columns])
```

**‚úîÔ∏è Verified: No Null Values Found**  


```python
# Total number of null values in the data
data.isnull().sum().sum() 
```

4. Remove Duplicates:

```python
data = data.drop_duplicates()
```

5. Convert Data Types

```python
# Convert to appropriate data types
data['Rank'] = data['Rank'].astype(int)
data['Members'] = data['Members'].astype(int)
data['Genres'] = data['Genres'].astype(str)

# Convert date columns to datetime objects
data['Aired'] = pd.to_datetime(data['Aired'], errors='coerce')
data['Premiered'] = pd.to_datetime(data['Premiered'], errors='coerce')
```

6. Feature Engineering

```python
data['Aired'] = data['Aired'].replace(np.nan, '1900-01-01')
data['Aired'].dt.year
```

```python
data['Premiered']=data['Premiered'].replace(np.nan, '2003-5-9')
data['Premiered'].dt.month
```

```python
# ADD column for: count the number of genres for each anime:
data['Num_Genres'] = data['Genres'].apply(lambda x: len(x.split(',')))
data['Num_Genres']
```

```python
# Create a binary column indicating if the anime has more than 50 episodes
data['Episodes'] = pd.to_numeric(data['Episodes'], errors='coerce')
data['More_Than_50_Episodes'] = data['Episodes'].apply(lambda x: 1 if (pd.notnull(x) and x > 50) else 0)
data['More_Than_50_Episodes']
```

```python
data.shape
```

```python
# Drop 2 columns we added
data = data.drop(columns=['More_Than_50_Episodes', 'Num_Genres'])
```

7. Normalize/Standardize Data

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
# Select numerical columns
numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns
# Initialize the scalers
scaler_standard = StandardScaler()
scaler_minmax = MinMaxScaler()
# Standardize the data in place
data[numerical_cols] = scaler_standard.fit_transform(data[numerical_cols])
# Normalize the data in place
data[numerical_cols] = scaler_minmax.fit_transform(data[numerical_cols])
# Display the modified DataFrame to verify
print("Standardized and Normalized Data")
print(data.head())
```

```python
data.describe()
```

## üîç Data Analysis  
Let's dive into our **Exploratory Data Analysis (EDA)** to reveal key insights! üìä  


```python
# Compute correlation matrix
correlation_matrix = data[['Score', 'Popularity', 'Rank', 'Members']].corr()
# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='RdPu', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

```python
palette = sns.color_palette("pastel6")
# Get the counts of each anime type
type_counts = data['Type'].value_counts()
# Plot the pie chart 
plt.figure(figsize=(8, 8))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', wedgeprops=dict(width=0.4), colors=palette)
plt.title('Distribution of Anime Types')
plt.axis('equal')
plt.show()
```

```python
# Plotting the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['Score'], data['Popularity'], alpha=0.5)
plt.title('Relationship between Score and Popularity')
plt.xlabel('Score')
plt.ylabel('Popularity')
plt.grid(True)
plt.show()
```

```python
# Split the 'Genres' column into multiple genres 
data['Genres'] = data['Genres'].str.split(',')
# Create a new DataFrame where each row corresponds to a single genre-score pair
genre_score_data = data.explode('Genres')[['Genres', 'Score']]
# Plotting
plt.figure(figsize=(12, 8))
sns.boxplot(x='Genres', y='Score', data=genre_score_data)
plt.title('Distribution of Scores across Genres')
plt.xlabel('Genres')
plt.ylabel('Score')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=data, x='Type', y='Members')
plt.xlabel('Type of Anime')
plt.ylabel('Members')
plt.title('Distribution of Members for Different Types of Anime (Violin Plot)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Assuming 'Aired' and 'Premiered' are datetime columns in DataFrame
# Plotting time series for 'Aired'
plt.figure(figsize=(12, 6))
sns.lineplot(x='Aired', y='Score', data=data, color='blue')
plt.title('Temporal Trends in Score (Aired)')
plt.xlabel('Aired Date')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

```python
# Counting the frequency of each genre
genre_counts = {}
for genre_list in data['Genres']:
    for genre in genre_list:
        genre_counts[genre] = genre_counts.get(genre, 0) + 1

# Sorting genres by frequency
sorted_genres = sorted(genre_counts.items(), key=lambda x: x[1], reverse=True)

# Creating a bar plot to visualize the frequency of each genre
plt.figure(figsize=(12, 6))
sns.barplot(x=[genre[1] for genre in sorted_genres], y=[genre[0] for genre in sorted_genres], palette='viridis')
plt.xlabel('Frequency')
plt.ylabel('Genres')
plt.title('Distribution of Anime Across Different Genres')
plt.show()
```

```python
# Create a box plot or violin plot to compare Scores or Popularity among different demographics
plt.figure(figsize=(12, 8))
sns.boxplot(x='Demographic', y='Score', data=data, palette='Set2')
plt.xlabel('Demographic')
plt.ylabel('Score')
plt.title('Distribution of Scores Across Different Demographics')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Plot the distribution of content ratings
plt.figure(figsize=(15, 3))
sns.countplot(data=data, x='Rating', palette='Set2')
plt.xlabel('Content Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Content Ratings')
plt.show()

# Analyze patterns in Scores based on ratings using box plots
plt.figure(figsize=(15, 3))
sns.boxplot(data=data, x='Rating', y='Score', palette='Set2')
plt.xlabel('Content Rating')
plt.ylabel('Score')
plt.title('Trends in Scores based on Ratings')
plt.show()
# Analyze trends in Popularity based on ratings using box plots
plt.figure(figsize=(15, 3))
sns.boxplot(data=data, x='Rating', y='Popularity', palette='Set2')
plt.xlabel('Content Rating')
plt.ylabel('Popularity')
plt.title('Trends in Popularity based on Ratings')
plt.show()
```

## üéØ Asking and Answering Questions  
1. How are the scores distributed among the top animes in the dataset?  


```python
# Filter the dataset to consider only the top animes 
top_animes = data[data['Rank'] <= 100]
# Create a histogram plot of the scores
plt.figure(figsize=(10,3))
sns.histplot(top_animes['Score'], kde=True, color='darkorchid', bins=20)
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Distribution of Scores Among Top Animes')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## üìä Popularity Distribution  
How is **anime popularity** spread across the dataset, and which titles rank among the most popular?  


```python
# Create a histogram of the popularity
plt.figure(figsize=(10, 3))
sns.histplot(data['Popularity'], kde=False, color='hotpink', bins=20)
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.title('Distribution of Anime Popularity')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
data['Genres'] = data['Genres'].astype(str)
# Sort the dataset by popularity
top_popular_animes = data.sort_values(by='Popularity', ascending=False).head(10)

# Display the top popular titles
print("Top 10 Most Popular Animes:")
print(top_popular_animes[['English', 'Popularity']])
```

### üìå Which 10 anime titles have the highest average ratings?


```python
top_10_average_ratings = data.groupby('English')['Score'].mean().sort_values(ascending=False).head(10)
#print(top_10_average_ratings)
plt.figure(figsize=(8, 8))
colors = ['pink','purple','hotpink','violet','lightpink','plum','deeppink','indigo','lightyellow','darkorchid']
explode=(0,0.2,0,0,0.2,0,0,0,0.1,0)   #to make a distance in selected bar
plt.pie(top_10_average_ratings,colors=colors,explode=explode,wedgeprops={'edgecolor':'black'},labels=top_10_average_ratings.index, autopct='%1.1f%%', startangle=50)
plt.title('Top 10 Average Ratings by Anime Title')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

What are the most common genres in the dataset, and which studios are responsible for producing the most animes?


```python
# Most prevalent genres
top_genres = data['Genres'].str.split(',').explode().str.strip().value_counts().head(10)
print("Top 10 Most Prevalent Genres:")
print(top_genres)
# Studios producing the most animes
top_studios = data['Studios'].str.split(',').explode().str.strip().value_counts().head(10)
print("\nTop 10 Studios Producing the Most Animes:")
print(top_studios)
```

What impact does the number of episodes have on an anime's average rating?


```python
# Group the data by the number of episodes and calculate the average rating
average_rating_by_episodes = data.groupby('Episodes')['Score'].mean()
# Plot the relationship
plt.figure(figsize=(10, 3))
plt.plot(average_rating_by_episodes.index, average_rating_by_episodes.values, marker='o', linestyle='-')
plt.xlabel('Number of Episodes')
plt.ylabel('Average Rating')
plt.title('Average Rating by Number of Episodes')
plt.grid(True)
plt.show()
```
